```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
y = ['Dog', 'Dog', 'Not Dog', 'Not Dog', 'Dog', 'Dog', 'Not Dog', 'Dog', 'Dog', 'Not Dog']
y_pred = ['Dog', 'Dog', 'Dog', 'Dog', 'Not Dog', 'Dog', 'Not Dog', 'Dog', 'Dog', 'Not Dog']
```

```python
accuracy_score(y, y_pred)
```

```python
confusion_matrix(y, y_pred)
```

```python
f1_score(y, y_pred, pos_label='Dog')
```

```python
print(classification_report(y, y_pred, target_names=['Dog', 'Not Dog']))
```
