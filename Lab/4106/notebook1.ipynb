```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to adjust brightness using loops and NumPy
def adjust_brightness(image, factor):
    height, width, channels = image.shape
    bright_image = np.zeros_like(image, dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            for c in range(channels):
                pixel = int(image[i, j, c] * factor)
                bright_image[i, j, c] = np.clip(pixel, 0, 255)
    return bright_image

# Load image
image = cv2.imread('sample.jpg')
if image is None:
    raise Exception("Image not found")

# Convert to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Adjust brightness
bright_image = adjust_brightness(image_rgb, 1.5)

# Convert to HSV and Grayscale
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Save processed images
cv2.imwrite('bright_image.jpg', cv2.cvtColor(bright_image, cv2.COLOR_RGB2BGR))
cv2.imwrite('hsv_image.jpg', cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR))
cv2.imwrite('gray_image.jpg', image_gray)

# Display using Matplotlib subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(image_rgb)
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')
axs[0, 1].imshow(bright_image)
axs[0, 1].set_title('Brightened Image')
axs[0, 1].axis('off')
axs[1, 0].imshow(image_hsv)
axs[1, 0].set_title('HSV Image')
axs[1, 0].axis('off')
axs[1, 1].imshow(image_gray, cmap='gray')
axs[1, 1].set_title('Grayscale Image')
axs[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
