```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Sine")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Cosine")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("input.jpg")

cv2.imshow("OpenCV BGR", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(img)
plt.title("Matplotlib shows BGR (Wrong Colors)")
plt.show()

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.title("Matplotlib shows RGB (Correct Colors)")
plt.show()
```
